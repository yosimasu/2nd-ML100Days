# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

def ReLU(x):
    return abs(x) * (x > 0)

x = plt.linspace(-1, 1, 100)

plt.plot(x, ReLU(x), 'b', label='linspace(-1, 1, 10)')

plt.grid()

plt.title('ReLU Function')

plt.show()
```

```python
def dReLU(x):
    return (1 * (x > 0))

x = plt.linspace(-1, 1, 100)

plt.plot(x, dReLU(x), 'b', label='linspace(-1, 1, 10)')

plt.grid()

plt.title('dReLU Function')

plt.show()
```
