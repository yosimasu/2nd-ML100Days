### 1. 模型的泛化能力 (generalization) 是指什麼？

    泛化能力是指機器學習對新樣本的適應能力，學習的目的在學到隱含在數據背後的規律，對具有同一規律的額外數據，經過訓練的模型也能給出合適的輸出。

### 2. 分類問題與回歸問題分別可用的目標函數有哪些？

* 分類問題
  * 交叉熵損失函數

* 回歸問題
  * 均方誤差
  * 平均絕對值誤差
