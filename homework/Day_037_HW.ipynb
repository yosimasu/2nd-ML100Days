  不行

2. 回歸模型是否對資料分布有基本假設?

  在使用迴歸分析前，必須要確認資料是否符合迴歸分析的基本統計假設，否則，當資料違反迴歸分析的基本統計假設時，會導致統計推論偏誤的發生。

  迴歸分析的基本統計假設有下列四項：
  
  * 線性關係
  * 常態性(normality) 
  * 誤差項的獨立性
  * 誤差項的變異數相等(Homoscedasticity)
